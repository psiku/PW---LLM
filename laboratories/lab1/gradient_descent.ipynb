# Zadanie : Napisać algorytm spadku wzdłuż gradientu (prosta implementacja)

### Używane biblioteki

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Funkcja i jej pochodna

```python
def f(x: float, y: float) -> float:
  return np.sin(x) * np.cos(y) + np.sin(0.5 *x) * np.cos(0.5 * y)

def f_grad(x: float, y: float) -> tuple:
  dx = np.cos(x) * np.cos(y) + 0.5 * np.cos(0.5 * x) * np.cos(0.5 * y)
  dy = -np.sin(x) * np.sin(y) - 0.5 * np.sin(0.5 * x) * np.sin(0.5 * y)
  return dx, dy
```

```python
x, y = 2, 2
print(f"x = {x}, y = {y}")
print(f"f(x,y) = {f(x,y)}")
print(f"f_grad(x,y) = {f_grad(x,y)}")
```

### Metoda spadku wzdłuż gradientu

```python
def make_step(x: float, y: float, learning_rate: float) -> tuple:
  dx, dy = f_grad(x, y)
  x -= learning_rate * dx
  y -= learning_rate * dy
  return x, y

def gradient_descent(x: float, y: float, learning_rate: float, epochs: int, verobse: bool=False) -> tuple:
    for _ in range(epochs):
        new_x, new_y = make_step(x, y, learning_rate)
        x, y = new_x, new_y

        print(f"x = {x}, y = {y}")
        print(f"f(x,y) = {f(x,y)}")

    return x, y
```

```python
starting_x, starting_y = 2, 2
learning_rate = 0.1
num_steps = 1000

x, y = gradient_descent(starting_x, starting_y, learning_rate, num_steps)
```
